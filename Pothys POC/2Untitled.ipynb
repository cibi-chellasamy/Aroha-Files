```python
stock_purchase_df=pd.DataFrame(columns=['stock_purchase_id','location', 'product_id', 'Stock opening date',
       'Stock closing date', 'Quantity purchase'])
```

```python
import datetime
from datetime import timedelta

def stock_opening_date():
    dates = list(pd.date_range('01-03-2024', '31-05-2024', freq='D'))
    return random.choice(dates)
def location():
    loc=list(range(1,17))
    return random.choice(loc)
def product():
    prod=list(range(100,1711))
    return random.choice(prod)

def quantity():
    qty=list(range(100,401,50))
    return random.choice(qty)
```

```python
stock_purchase_df
```

```python
for i in range(10000):
    stock_purchase_df.loc[i, 'location'] = location()  
    stock_purchase_df.loc[i, 'product_id'] = product() 
    stock_purchase_df.loc[i, 'Stock opening date'] = stock_opening_date() 
    stock_purchase_df.loc[i, 'Stock closing date'] = stock_purchase_df.loc[i, 'Stock opening date'] + timedelta(days=120)
    stock_purchase_df.loc[i, 'Quantity purchase'] = quantity()
```

```python
sales_fact=pd.DataFrame(columns=['sales_id','sales_date','stock_id','Location','Product_id','quantity_sold','Price'])
```

```python
def stock_id():
    stock=list(stock_purchase_df['stock_purchase_id'])
    stock_id = random.choice(stock)
    return stock_id

def get_price(stock_id):
    prod = stock_purchase_df.loc[stock_purchase_df['stock_purchase_id'] == stock_id, 'product_id'].values[0]
    price=product_df.loc[product_df['product_id']==prod,'MRP Price (With Tax)'].values[0]
    return price

def loc_sales(stock_id):
    loc = stock_purchase_df.loc[stock_purchase_df['stock_purchase_id'] == stock_id, 'location'].values[0]
    return loc

def get_product(stock_id):
    prod = stock_purchase_df.loc[stock_purchase_df['stock_purchase_id'] == stock_id, 'product_id'].values[0]
    return prod

def sales_date(stock_id):
    stock_open_date = stock_purchase_df.loc[stock_purchase_df['stock_purchase_id'] == stock_id, 'Stock opening date'].values[0]
    dates = pd.date_range(stock_open_date, pd.Timestamp.now())
    date = random.choice(dates)
    return date

def quantity_sold(stock_id):
    pur_qty=stock_purchase_df.loc[stock_purchase_df['stock_purchase_id'] == stock_id, 'Quantity purchase'].values[0]
    df=sales_fact[sales_fact['stock_id']==stock_id]
    sum_qty=df.groupby('stock_id')['quantity_sold'].sum().values[0]
    if sum_qty<pur_qty:
        qty_=[1,2,3]
        ind=(pur_qty-sum_qty)-1
        if ind>1:
            qty_=qty_[:ind]
            qty=random.choice(qty_)
        else:
            qty=0
    else:
        qty=0
    return qty


# def quantity_sold(stock_id):
#     pur_qty = stock_purchase_df.loc[stock_purchase_df['stock_purchase_id'] == stock_id, 'Quantity purchase'].values[0]
#     df = sales_fact[sales_fact['stock_id'] == stock_id]
#     sum_qty = df['quantity_sold'].sum()

#     if sum_qty < pur_qty:
#         qty = random.choice([1, 2, 3])
#     else:
#         qty = 0

#     return qty
```

```python
for i in range(346540):
    sales_fact.loc[i, 'stock_id'] = stock_id()
    sales_fact.loc[i, 'sales_date'] = sales_date(sales_fact.loc[i, 'stock_id'])
    sales_fact.loc[i, 'Location'] = loc_sales(sales_fact.loc[i, 'stock_id'])
    sales_fact.loc[i, 'Product_id'] = get_product(sales_fact.loc[i, 'stock_id'])
    sales_fact.loc[i, 'quantity_sold'] = quantity_sold(sales_fact.loc[i, 'stock_id'])
    sales_fact.loc[i, 'Price'] = get_price(sales_fact.loc[i, 'stock_id'])
```
