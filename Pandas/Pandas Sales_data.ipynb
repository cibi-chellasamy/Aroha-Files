```python
# Create a new column in the sales_data DataFrame named 'TotalPrice' by multiplying 'Quantity' 
# and 'Price'.
sales_df['Total_price'] = sales_df['Quantity']*sales_df['Price']
sales_df
```

```python
#. If the 'Date' column in sales_data is not in datetime format, convert it to datetime.

sales_df['Date'] = pd.to_datetime(sales_df['Date'])
sales_df.dtypes
```

```python
#. Sort the sales_data DataFrame by 'Date' in ascending order

sales_df = sales_df.sort_values(by = 'Date',ascending=True)
```

```python
sales_df
```

```python
#Get the unique values in the 'Product' column of the sales_data DataFrame.

sales_df['Product'].unique()
```

```python
#Create a pivot table from the sales_data DataFrame, where rows represent 'Product' and columns 
#represent 'Date', and the values are the total quantity sold.

pivot_tab= pd.pivot_table(sales_df, values='Quantity', index='Product', columns='Date', aggfunc='sum',fill_value=0)
```

```python
pivot_tab
```

```python
#. Increase the 'Price' by 10% for all products with a total quantity sold greater than 100.

tota_quan_sold = sales_df.groupby('Product')['Quantity'].sum()

prod_great_100 = tota_quan_sold[tota_quan_sold>100]
```

```python
tota_quan_sold
```

```python
prod_great_100
```

```python
sales_df.loc[sales_df['Product'].isin(prod_great_100), 'Price'] *= 1.1
```

```python
sales_df
```

```python
#. Generate a statistical summary of the sales_data DataFrame, including mean, median, and standard deviation

sales_df.describe()
```

```python
# If the 'Date' column is in datetime format, resample the data to get the monthly total sales.

monthly_total_sales = sales_df.resample('M').sum()
```

```python
#Create a new column in the sales_data DataFrame named 'CumulativeSales' representing the cumulative sum of 'Quantity' for each product
sales_df['cumulativeSales'] = sales_df.groupby('Product')['Quantity'].cumsum()
```

```python
sales_df
```

```python
total_sales = sales_df.groupby('Product')['Quantity'].sum()
```

```python
products_great_50 = total_sales[total_sales > 50]
products_great_50
```

```python
avg_price = sales_df[sales_df['Product'].isin(products_great_50)].groupby('Product')['Price'].mean()
avg_price
```
