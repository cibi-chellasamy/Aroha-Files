```python
df_finance[['First_name','Last_name']] = df_finance['Patient_Name'].str.split(',',expand = True)
df_finance.head(2)
```

```python
df_finance = df_finance[['Patient_Name', 'First_name', 'Last_name', 'Patient_Number', 'Service_Date', 'Logbook_Status',
       'Clinic', 'Discharging_Provider', 'Financial_Class', 'Payer_Name',
       'RTE_Resonse', 'CardOnFile', 'Card_Level', 'Fee_Schedule',
       'Charged_Amount', 'CopayUpfront_Payments', 'Additional_Payments',
       'Other_Payments', 'Adjustments', 'Remaining_Balance',
       'ProcedureCodes_Charges']]
```

```python
df_finance.drop('Patient_Name',axis= 1,inplace=True)
```

```python
def extract_procedure(procedure_str):
    procedures = procedure_str.split('|')
    procedures = [proc.strip() for proc in procedures] 
    procedure_codes = []
    charges = []

    for proc in procedures:
        code, charge = proc.split(':')
        procedure_codes.append(code.strip())
        charges.append(charge.strip())
    
    return pd.Series([procedure_codes, charges])



df_finance[['Procedure_Codes', 'Charges']] = df_finance['ProcedureCodes_Charges'].apply(extract_procedure)
```

```python
df_finance.head(2)
```

```python
df_finance['Charges'] = df_finance['Charges'].apply(lambda x: [float(charge.replace('$', '')) for charge in x])
```

```python
df_finance.rename(columns = {'Charges':'Charges_in_$'},inplace=True)
```
