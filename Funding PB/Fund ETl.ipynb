```python
def print_null_shapes(dataframes_dict):
    for name, df in dataframes_dict.items():
        null_counts = df.isnull().sum()

        print(f"{name}")
        print("Shape of DataFrame:")
        print(df.shape)
        print("Null value counts per column:")
        print(null_counts)
        print("Data Types")
        print(df.dtypes)
        print("\n" + "="*50 + "\n")

dataframes_dict = {
    'donor_df': donor_df,
    'gift_df': gift_df,
    'gift_fund_map_df': gift_fund_map_df,
    'department_df': department_df,
    'fund_df': fund_df,
    'term_df': term_df,
    'purpose_df': purpose_df,
    'lookup_tables_df': lookup_tables_df,
    'asset_management_company_df': asset_management_company_df,
    'fund_allocation_df': fund_allocation_df,
    'fund_returns_df': fund_returns_df,
    'fund_value_history_df': fund_value_history_df,
    'fund_sector_df': fund_sector_df,
    'fund_disbursement_df': fund_disbursement_df
}

print_null_shapes(dataframes_dict)
```
