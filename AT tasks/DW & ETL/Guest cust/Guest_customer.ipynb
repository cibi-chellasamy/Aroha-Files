```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df_cust = pd.read_excel('Customer.xlsx')
df_guest = pd.read_excel('Guest.xlsx')
```

```python
df_guest.head(10)
```

```python
df_cust.head(10)
```

```python
df_call_plan = pd.DataFrame({
    'Call_plan_id': [], 
    'Guest_name': [], 
    'Guest_city': [], 
    'Guest_phone': []
})
```

```python
for index, guest_row in df_guest.iterrows():
    guest_name = guest_row['name']
    guest_phone = guest_row['phone']
    guest_city = guest_row['city']
    
    match = ((df_cust['c_name'] == guest_name)&
            (df_cust['c_phone']==guest_phone)&
            (df_cust['c_city']==guest_city)).any()
    if match:
        df_guest.drop(index,inplace=True)
    else:
        df_call_plan = df_call_plan.append({'Call_plan_id': len(df_call_plan)+1, 
                                            'Guest_name': guest_name, 
                                            'Guest_city': guest_city, 
                                            'Guest_phone': guest_phone}, 
                                           ignore_index=True)
df_guest['pro_flag'] = 'y'
pd.set_option('display.float_format',lambda x:'%.0f'%x)
```
