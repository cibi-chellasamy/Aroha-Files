```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df_cust = pd.read_excel('Customer.xlsx')
df_guest = pd.read_excel('Guest.xlsx')
```

```python
df_call_plan = pd.DataFrame({
    'Call_plan_id': [], 
    'Guest_name': [], 
    'Guest_city': [], 
    'Guest_phone': []
})
```

```python
df_existing_guests = df_guest.merge(df_cust, how='inner', 
                                    left_on=['name', 'phone', 'city'], 
                                    right_on=['c_name', 'c_phone', 'c_city'])
```

```python
df_guest = df_guest[~df_guest.index.isin(df_existing_guests.index)]
```

```python
df_call_plan = df_guest[['name', 'phone', 'city']].copy()
df_call_plan.rename(columns=
                    {'name': 'Guest_name', 'phone': 'Guest_phone', 'city': 'Guest_city'}, 
                    inplace=True)
```
