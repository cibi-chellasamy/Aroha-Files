```python
import pandas as pd
```

```python
df = pd.read_excel('market equipment data of HealthCare V1 - Copy.xlsx')
```

```python
df.head(2)
```

```python
df_h = pd.DataFrame(df[['businessname','practicename','practiceaddress1','practiceaddress2','practicecity','practicestate','practicezip9code','telephonenumber']])
```

```python
df_h.head(2)
```

```python
df_hm = pd.read_excel('Hospital Master xl.xlsx')
```

```python
df_hm
```

```python
df_branch = pd.merge(left=df_hm, right=df_h, left_on='businessname', right_on='businessname')
df_branch = df_branch.drop(['businessname'], axis=1)  
```

```python
df_branch.drop('participationbegindate',inplace=True,axis=1)
df_branch.drop('practicename_x',inplace=True,axis=1)
```

```python
df_h.head(2)
```

```python
df_branch['Branch_Id'] = range(1500, 1500+int(len(df_branch)))
```
