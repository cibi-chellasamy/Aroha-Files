```python
data = {
    'CONTRACT_ID': [12345, 12879, 12987, 12346, 12348],
    'CONTRACT_TYPE': ['government', 'individual', 'individual', 'government', 'government'],
    'CONT_S_DATE': ['10-Jan-10', '03-FEB-11', '15-Feb-11', '13-MAR-09', '6-Apr-12'],
    'CONT_E_DATE': ['10-JAN-12', '15-Feb-12', '15-SEPT-11', '14-MAR-11', '22-May-13'],
    'CPT_ID': [100, 101, 100, 102, 103, ],
    'CONT_AMT': [700000, 489938, 289456, 200000, 200000],
    'INS_DATE': ['10-Jan-10', '3-Feb-11', '3-Feb-11', '13-Mar-09', '6-Apr-12'],
    'UPD_DATE': [np.nan, np.nan, '10-Apr-09', np.nan,np.nan ]
}


df_contract = pd.DataFrame(data)
df_contract['CONT_S_DATE'] = pd.to_datetime(df_contract['CONT_S_DATE'])
df_contract['CONT_E_DATE'] = pd.to_datetime(df_contract['CONT_E_DATE'])
df_contract['INS_DATE'] = pd.to_datetime(df_contract['INS_DATE'])
df_contract['UPD_DATE'] = pd.to_datetime(df_contract['UPD_DATE'])
df_contract
```

```python
df_contract.CONT_S_DATE
```

```python

result=pd.DataFrame()
for index,rows in df_contract.iterrows():
    s_date=rows['CONT_S_DATE']
    e_date=rows['CONT_E_DATE']
    due=rows['CPT_ID']
    amt=rows['CONT_AMT']
    id=rows['CONTRACT_ID']
    day=rows.INS_DATE.day
    mon=rows.INS_DATE.month
    df=pd.DataFrame()
    re=pd.DataFrame()
    date_range = pd.date_range(s_date, e_date - timedelta(days = 1), freq='d')
    date_df = pd.DataFrame({'start_date':[ s_date],'end_date': [e_date]}).apply(pd.to_datetime)
    if due==100:
        date_df = pd.DataFrame({'start_date':[ s_date],'end_date': [e_date]}).apply(pd.to_datetime)
        date_df['months_diff'] = (date_df['end_date'] - date_df['start_date']) / np.timedelta64(1, 'D')
        month=date_df['months_diff']/365*(12)
        month=round(month)
        print(month)
        amt=(amt/month)
        amt_mon=amt
        df['date']=date_range
        date=df.loc[df['date'].dt.day == day].reset_index(drop=True) 
        re['date']=date
        re['amt']=amt_mon
        re['cont_id']=id
        result=pd.concat([result,re])
    if due==101:
        mon=rows.INS_DATE.month
        date_df = pd.DataFrame({'start_date':[ s_date],'end_date': [e_date]}).apply(pd.to_datetime)
        date_df['months_diff'] = (date_df['end_date'] - date_df['start_date']) / np.timedelta64(1, 'D')
        month=date_df['months_diff']/365*(12/4)
        amt=amt/month
        df['date']=date_range
        date=df.loc[df['date'].dt.day == day].reset_index(drop=True)
        d=date.loc[(date['date'].dt.month== mon)|(date['date'].dt.month== mon+3)|(date['date'].dt.month== mon+6)|(date['date'].dt.month== mon+9)]
        re['date']=d
        re['amt']=amt
        re['cont_id']=id
        display(re)
        result=pd.concat([result,re])
    elif due==102:
        mon=rows.INS_DATE.month
        date_df = pd.DataFrame({'start_date':[ s_date],'end_date': [e_date]}).apply(pd.to_datetime)
        date_df['months_diff'] = (date_df['end_date'] - date_df['start_date']) / np.timedelta64(1, 'D')
        month=date_df['months_diff']/365*(12/6)
        amt=amt/month
        # print(amt)
        df['date']=date_range
        date=df.loc[df['date'].dt.day == day].reset_index(drop=True)
        d=date.loc[(date['date'].dt.month== mon)|(date['date'].dt.month== mon+3)]
        re['date']=d
        re['amt']=amt
        re['cont_id']=id
        print(re)
        result=pd.concat([result,re])
    elif due==103:
        mon=rows.INS_DATE.month
        date_df = pd.DataFrame({'start_date':[ s_date],'end_date': [e_date]}).apply(pd.to_datetime)
        date_df['date_diff'] = (date_df['end_date'] - date_df['start_date']) / np.timedelta64(1, 'D')
        month=date_df['date_diff']/365*(12/12)
        month=round(month)
        print(month)
        amt=amt/month
        amount=amt
        df['date']=date_range
        res=df.loc[df['date'].dt.day == day].reset_index(drop=True)
        res=res.loc[res['date'].dt.month == mon].reset_index(drop=True)
        re['date']=res
        re['amt']=amount
        re['cont_id']=id
        result=pd.concat([result,re])
    
    else:
        continue
```
