```python
import pandas as pd
import glob
import os
import json
from datetime import datetime
```

```python
file_paths = glob.glob(r'C:\Users\DELL\OneDrive\Desktop\AT files\Daily\raw\*.csv')

circuit_df = pd.DataFrame()
races_df = pd.DataFrame()

for file_path in file_paths:
    filename = os.path.basename(file_path).split('.')[0]
    df = pd.read_csv(file_path)
    if 'circuits' in filename.lower():
        circuit_df = df
    elif 'races' in filename.lower():
        races_df = df
```

```python
directory = r'C:\Users\DELL\OneDrive\Desktop\AT files\Daily\raw'
file_paths = {
    'results': os.path.join(directory, 'results.json'),
    'pit_stops': os.path.join(directory, 'pit_stops.json'),
    'drivers': os.path.join(directory, 'drivers.json'),
    'constructors': os.path.join(directory, 'constructors.json')
}
```

```python
results_df = pd.read_json(file_paths['results'],lines=True)
constructors_df = pd.read_json(file_paths['constructors'],lines=True)
pit_stops_df = pd.read_json(file_paths['pit_stops'])
drivers_df = pd.read_json(file_paths['drivers'],lines=True)
```

```python
results_df 
constructors_df 
pit_stops_df
drivers_df
circuit_df
races_df
```

```python
dataframes = [results_df, constructors_df, pit_stops_df, drivers_df, circuit_df, races_df]
for df in dataframes:
    if 'url' in df.columns:
        df.drop(columns='url', inplace=True)
```

```python
today_date = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
dataframes = [results_df, constructors_df, pit_stops_df, drivers_df, circuit_df, races_df]
for df in dataframes:
    df['created_date'] = today_date
```

```python
column_mappings = {'raceId': 'race_id','circuitId':'circuit_id','driverId':'driver_id','raceID':'race_id','constructorId':'constructor_id','resultId':'result_id'}
dataframes = [results_df, constructors_df, pit_stops_df, drivers_df, circuit_df, races_df]

for df in dataframes:
    for current_col, new_col in column_mappings.items():
        if current_col in df.columns:
            df.rename(columns={current_col: new_col}, inplace=True)
```
