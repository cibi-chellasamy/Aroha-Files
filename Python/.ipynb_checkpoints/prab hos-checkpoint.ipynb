```python
import pandas as pd
pf=pd.read_excel('market equipment data of HealthCare.xlsx')
```

```python
pf
```

```python
p=pd.DataFrame(pf[['businessname','practicename','participationbegindate']])
p
```

```python
uniq=p.drop_duplicates()
uniq = uniq.groupby(by=["businessname",'participationbegindate'])['practicename'].unique().reset_index()
uniq
```

```python
id=len(uniq['businessname'])
id
```

```python
hospital_id=[]
for i in range(id):
    idn=500+i
    hospital_id.append(idn)
hospital_id    
    
```

```python
hospital_master=pd.DataFrame()
hospital_master['hospital_id']=hospital_id
hospital_master['hospital_name']=uniq['businessname']
hospital_master['practice_name']=uniq['practicename']
hospital_master['create_date']=uniq['participationbegindate']
hospital_master
```

```python
p=pd.DataFrame(pf[['businessname','practicename','practiceaddress1','practiceaddress2','practicecity','practicestate','practicezip9code','telephonenumber']])
p
```

```python
branch=pd.merge(left=hospital_master[['hospital_name','hospital_id']],right=p,left_on=hospital_master['hospital_name'],right_on=p['businessname'])\
.drop(['key_0','hospital_name','businessname'],axis=1,)
branch
```

```python
id=len(branch['practicename'])
branch_id=[]
for i in range(id):
    idn=1500+i
    branch_id.append(idn)
branch_id  
```

```python
branch_m=pd.DataFrame()
branch_m['branch_id']=branch_id
branch_m['hospital_id']=branch['hospital_id']
branch_m['practicename']=branch['practicename']
branch_m['practiceaddress1']=branch['practiceaddress1']
branch_m['practiceaddress2']=branch['practiceaddress2']
branch_m['practicecity']=branch['practicecity']
branch_m['practicestate']=branch['practicestate']
branch_m['practicezip9code']=branch['practicezip9code']
branch_m['telephonenumber']=branch['telephonenumber']

branch_m
```
